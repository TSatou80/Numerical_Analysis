```python
from math import sin
from numpy import array,arange
from pylab import plot,xlabel,show
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = 0.0
b = 20
N = 5000
h = (b-a)/N

si = 10
rr = 28
be = 8/3

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

def f(r,t):
    x = r[0]
    y = r[1]
    z = r[2]
    fx = si*(y - x)
    fy = rr*x - y - x*z
    fz = x*y - be*z
    return array([fx,fy,fz], float)

tpoints = arange(a,b,h)
xpoints = []
ypoints = []
zpoints = []

r = array([0,1,0],float)
for t in tpoints:
    xpoints.append(r[0])
    ypoints.append(r[1])
    zpoints.append(r[2])
    k1 = h*f(r,t)
    k2 = h*f(r+0.5*k1, t+0.5*h)
    k3 = h*f(r+0.5*k2, t+0.5*h)
    k4 = h*f(r+k3, t+h)
    r += (k1+2*k2+2*k3+k4)/6

# plot(tpoints, ypoints)
plot(xpoints, ypoints, zpoints)
```
