```python
from math import sin,cos
from numpy import array,arange,pi
from pylab import plot,xlabel,show
```

```python
m = 9.1094e-31
hbar = 1.0546e-34
e = 1.6022e-19
L = 5.2918e-11
N = 1000
h = L/N
a = 1.0e-11

def V(x):
    V0 = 50*e
    return V0*x**2/a**2

def f(r,x,E):
    psi = r[0]
    phi = r[1]
    fpsi = phi
    fphi = (2*m/hbar**2)*(V(x) - E)*psi
    return array([fpsi,fphi],float)

xpoints = arange(-5*a,5*a,h)
energy = [138*e,414*e,690*e]

for i in energy:
    psipoints = []
    phipoints = []
    r = array([0,1],float)
    for x in xpoints:
        psipoints.append(r[0])
        phipoints.append(r[1])
        k1 = h*f(r, x, i)
        k2 = h*f(r+0.5*k1, x+0.5*h, i)
        k3 = h*f(r+0.5*k2, x+0.5*h, i)
        k4 = h*f(r+k3, x+h, i)
        r += (k1 + 2*k2 + 2*k3 + k4)/6
    plot(xpoints, psipoints)
    #plot(xpoints, phipoints)
```
