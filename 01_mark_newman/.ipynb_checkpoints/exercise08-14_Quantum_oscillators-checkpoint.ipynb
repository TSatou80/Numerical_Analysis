# Exercise 8.14 Quantum oscillators (p397)

```python
from numpy import array,arange
from math import sin, cos
from numpy import array,arange, pi
from pylab import plot, xlabel, show, grid

m = 9.1094e-31
hbar = 1.0546e-34
e = 1.6022e-19
a = 5.2918e-11
L = a*3
N = 1000
h = L/N
V0 = 50*e

def V(x):    
    return V0*x**2/((1*a)**2)

def f(r,x,E):
    psi = r[0]
    phi = r[1]
    fpsi = phi
    fphi = (2*m/hbar**2)*(V(x) - E)*psi
    return array([fpsi, fphi], float)


```

```python
def solve(E):
    psi = 0.0
    phi = 1.0
    r = array([psi,phi],float)
    
    for x in arange(-L/2,L/2,h):
        k1 = h*f(r,x,E)
        k2 = h*f(r+0.5*k1, x+0.5*h, E)
        k3 = h*f(r+0.5*k2, x+0.5*h, E)
        k4 = h*f(r+k3, x+h, E)
        r += (k1 + 2*k2 + 2*k3 + k4)/6
#     print(r[0],r[1]) 
    
    return r[0]
```

```python
#-----------------------
E1 = 10*e
E2 = 40*e
E = 0
psi2 = solve(E1)
target = e/1000
while abs(E1 - E2) > target:
    psi1, psi2 = psi2, solve(E2)
    E1, E2 = E2,E2 - psi2*(E2 - E1)/(psi2 - psi1)

E = E2   
print("E=", E2/e, "eV")

E = 27.54762784529952*e

psi = 0.0
phi = 1.0

tpoints = arange(-L/2,L/2,h)
xpoints1 = []

r = array([psi,phi],float)
for x in tpoints:
    k1 = h*f(r,x,E)
    k2 = h*f(r+0.5*k1, x+0.5*h, E)
    k3 = h*f(r+0.5*k2, x+0.5*h, E)
    k4 = h*f(r+k3, x+h, E)
    r += (k1 + 2*k2 + 2*k3 + k4)/6
    xpoints1.append(r[0])

total1 = sum((tpoints*xpoints1)**2)
    
#-----------------------    
E1 = 40*e
E2 = 80*e
E = 0
psi2 = solve(E1)
target = e/1000
while abs(E1 - E2) > target:
    psi1, psi2 = psi2, solve(E2)
    E1, E2 = E2,E2 - psi2*(E2 - E1)/(psi2 - psi1)

E = E2   
print("E=", E2/e, "eV")

psi = 0.0
phi = 1.0

tpoints = arange(-L/2,L/2,h)
xpoints2 = []

r = array([psi,phi],float)
for x in tpoints:
    k1 = h*f(r,x,E)
    k2 = h*f(r+0.5*k1, x+0.5*h, E)
    k3 = h*f(r+0.5*k2, x+0.5*h, E)
    k4 = h*f(r+k3, x+h, E)
    r += (k1 + 2*k2 + 2*k3 + k4)/6
    xpoints2.append(r[0])

total2 = sum((tpoints*xpoints2)**2)


#-----------------------
E1 = 90*e
E2 = 170*e
E = 0
psi2 = solve(E1)
target = e/1000
while abs(E1 - E2) > target:
    psi1, psi2 = psi2, solve(E2)
    E1, E2 = E2,E2 - psi2*(E2 - E1)/(psi2 - psi1)

E = E2   
print("E=", E2/e, "eV")

psi = 0.0
phi = 1.0

tpoints = arange(-L/2,L/2,h)
xpoints3 = []

r = array([psi,phi],float)
for x in tpoints:
    k1 = h*f(r,x,E)
    k2 = h*f(r+0.5*k1, x+0.5*h, E)
    k3 = h*f(r+0.5*k2, x+0.5*h, E)
    k4 = h*f(r+k3, x+h, E)
    r += (k1 + 2*k2 + 2*k3 + k4)/6
    xpoints3.append(r[0])

total3 = sum((tpoints*xpoints3)**2)
    
    
#-----------------------
E1 = 170*e
E2 = 280*e
E = 0
psi2 = solve(E1)
target = e/1000
while abs(E1 - E2) > target:
    psi1, psi2 = psi2, solve(E2)
    E1, E2 = E2,E2 - psi2*(E2 - E1)/(psi2 - psi1)

E = E2   
print("E=", E2/e, "eV")

psi = 0.0
phi = 1.0

tpoints = arange(-L/2,L/2,h)
xpoints4 = []

r = array([psi,phi],float)
for x in tpoints:
    k1 = h*f(r,x,E)
    k2 = h*f(r+0.5*k1, x+0.5*h, E)
    k3 = h*f(r+0.5*k2, x+0.5*h, E)
    k4 = h*f(r+k3, x+h, E)
    r += (k1 + 2*k2 + 2*k3 + k4)/6
    xpoints4.append(r[0])

total4 = sum((tpoints*xpoints4)**2)


    
grid()
plot(tpoints, xpoints1/total1)
plot(tpoints, xpoints2/total2)
plot(tpoints, xpoints3/total3)
plot(tpoints, xpoints4/total4)
```
