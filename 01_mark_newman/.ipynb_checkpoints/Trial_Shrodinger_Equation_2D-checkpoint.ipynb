```python
%matplotlib notebook
import numpy as np
from numpy import empty, zeros, max, ones, array, arange, pi
from pylab import imshow, gray, show, imshow
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from math import sin, cos
from pylab import plot, xlabel, show, grid
# #タイトルで漢字が使えるようフォントを設定
# plt.rcParams['font.family'] = 'Meiryo'
M = 100
V = 1.0
target = 1e-3
m = 9.1094e-31
hbar = 1.0546e-34
e = 1.6022e-19
L = 5.2918e-11
N = 1000
h = L/N
V0 = 100*e
```

```python
#create arrays to hold potential values
phi = zeros([M+1,M+1],float)
# for i in range(60,80):
#     for j in range(20,40):
#         phi[i,j] = 5

# for i in range(20,40):
#     for j in range(60,80):
#         phi[i,j] = -5
        
phi[0,:] = 0
phi[M,:] = 0
phi[:,0] = 0
phi[:,M] = 0

count = 0

phiprime = empty([M+1,M+1],float)

#main loop
delta = 1.0
while delta > target:
    #calculate new values of the potential
    for i in range(M+1):
        for j in range(M+1):
            if i==0 or i==M or j==0 or j==M:
                phiprime[i,j] = 0
            else:
                 phiprime[i,j] = (phi[i+1,j] + phi[i-1,j] + phi[i,j+1] + phi[i,j-1])/4 + 0.1
    
    #calculate maximum difference from old values
    delta = max(abs(phi - phiprime))
    
    #swap the two arrays around
    phi, phiprime = phiprime, phi
    count += 1
#     print(phi)
    if count > 600:
        break
            
# imshow(phi, cmap="bwr")
# gray()
show()
```

```python
def V(x):    
    return V0*(x/L)*(x/L-1) 

# def V(x):
#     return 0.0


def f(r,x,E):
    psi = r[0]
    phi = r[1]
    fpsi = phi
    fphi = (2*m/hbar**2)*(V(x) - E)*psi
    return array([fpsi, fphi], float)
```

```python
def solve(E):
    psi = 0.0
    phi = 1.0
    r = array([psi,phi],float)
    
    for x in arange(0,L,h):
        k1 = h*f(r,x,E)
        k2 = h*f(r+0.5*k1, x+0.5*h, E)
        k3 = h*f(r+0.5*k2, x+0.5*h, E)
        k4 = h*f(r+k3, x+h, E)
        r += (k1 + 2*k2 + 2*k3 + k4)/6
#     print(r[0],r[1]) 
    
    return r[0]
```

```python
E1 = 1000*e
E2 = 2000*e
psi2 = solve(E1)
target = e/1000
while abs(E1 - E2) > target:
    psi1, psi2 = psi2, solve(E2)
    E1, E2 = E2,E2 - psi2*(E2 - E1)/(psi2 - psi1)
#     print(E2/e)
    
print("E=", E2/e, "eV")
```

### 波動関数計算用
- 定常状態のエネルギーを代入して、波動関数を求める。


```python
E1 = 112.5*e
E2 = 519.2*e
E3 = 1191.4*e
E4 = 2131.6*e
```

```python
E = E1

psi = 0.0
phi = 1.0

tpoints = arange(0,L,h)
xpoints = []

r = array([psi,phi],float)
for x in arange(0,L,h):
    k1 = h*f(r,x,E)
    k2 = h*f(r+0.5*k1, x+0.5*h, E)
    k3 = h*f(r+0.5*k2, x+0.5*h, E)
    k4 = h*f(r+k3, x+h, E)
    r += (k1 + 2*k2 + 2*k3 + k4)/6
    xpoints.append(r[0])

grid()
plot(tpoints, xpoints)
```

```python
E = E2

psi = 0.0
phi = 1.0

tpoints = arange(0,L,h)
xpoints2 = []

r = array([psi,phi],float)
for x in arange(0,L,h):
    k1 = h*f(r,x,E)
    k2 = h*f(r+0.5*k1, x+0.5*h, E)
    k3 = h*f(r+0.5*k2, x+0.5*h, E)
    k4 = h*f(r+k3, x+h, E)
    r += (k1 + 2*k2 + 2*k3 + k4)/6
    xpoints2.append(r[0])

grid()
plot(tpoints, xpoints2)
```

### x方向とy方向を独立に算出して、足し合わせる

```python
xx = np.matrix([xpoints,])
xx2 = np.matrix([xpoints2,])

for i in range(len(xpoints)-1):
    xx = np.insert(xx,xx.shape[0],xpoints,axis=0)

for i in range(len(xpoints2)-1):
    xx2 = np.insert(xx2,xx2.shape[0],xpoints2,axis=0)

# tmp = np.power(xx,2).T + np.power(xx2,2)
# tmp = xx.T + xx2
tmp = np.multiply(xx.T, xx2)

# imshow(tmp)

X, Y = np.meshgrid(tpoints,tpoints)
```
