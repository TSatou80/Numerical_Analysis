```python
from math import pi
from numpy import linspace,array,cos,sin,exp
#numpyのsin,cosをインポート mathではだめ。
from matplotlib.pyplot import plot
```

```python
theta = linspace(0,24*pi,1000)
r = exp(cos(theta)) - 2*cos(4*theta) + sin(theta/12)**5
```

```python
x = r*cos(theta)
y = r*sin(theta)
#この部分で、mathのsin,cosは使えない
```

```python
plot(x,y)
```
