```python
from numpy import pi,cos,sin,array,empty,zeros
from math import sqrt
from matplotlib.pyplot import plot,scatter, imshow, colorbar,gray,hot,xlim,ylim
```

```python
def f(theta, x=0, m=0):
    return cos( m*theta - x*sin(theta) )
```

```python
N = 100
a = 0
b = pi
h = (b-a)/N
```

```python
def Jm(x_=0, m_=0):
    total = 0
    total += ( f(a, x=x_, m=m_) + f(b, x=x_, m=m_) )
    for i in range(1,int(N),2):
        total += 4*f(a + i*h, x=x_, m=m_)
    for i in range(2,int(N),2):
        total += 2*f(a + i*h, x=x_, m=m_)
    return total*h/3
```

```python
from numpy import linspace
```

```python
X = linspace(0,20,100)
```

```python
from pylab import plot
```

```python
for i in range(3):
    plot(X,Jm(x_=X, m_=i))
```

```python
x = linspace(6e-7,1e-5,200)
y = linspace(6e-7,1e-5,200)
```

```python
R = empty([x.size,y.size]) 
```

```python
for i in range(0,x.size):
    for j in range(0,y.size):
        R[i,j] = sqrt(x[i]**2 + y[j]**2)
```

```python
I = empty([x.size,y.size]) 
```

```python
lam = 5e-6 
k = 2*pi/lam
for i in range(x.size):
    for j in range(y.size):
        I[i][j] = (Jm(k*R[i][j],m_=0)/(k*R[i][j]))**2
```

```python
imshow(I,vmax=0.07)
colorbar()
# xlim(90,110)
# ylim(90,110)
```
