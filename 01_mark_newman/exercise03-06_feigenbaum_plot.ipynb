```python
from matplotlib.pyplot import plot,xlabel, ylabel
from numpy import linspace
```

```python
x = 0.5
y = 0
r = linspace(1,4,1000)
x_result = []
xx = []
xx.append(x)
for r_ in r:
    for i in range(100):
        y = r_*x*(1-x)
        xx.append(y)
        x = y
    x_result.append(y)

```

```python
plot(r,x_result)
xlabel("r")
ylabel("x")
```
