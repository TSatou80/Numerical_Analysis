```python
from math import sin
from numpy import array,arange
from pylab import plot,xlabel,show
```

```python
a = 0.0
b = 20
N = 600
h = (b-a)/N

al = 1
be = 0.5
ganma = 0.5
de = 2

def f(r,t):
    x = r[0]
    y = r[1]
    fx = al*x - be*x*y
    fy = ganma*x*y - de*y
    return array([fx,fy], float)

tpoints = arange(a,b,h)
xpoints = []
ypoints = []

r = array([12,20],float)
for t in tpoints:
    xpoints.append(r[0])
    ypoints.append(r[1])
    k1 = h*f(r,t)
    k2 = h*f(r+0.5*k1, t+0.5*h)
    k3 = h*f(r+0.5*k2, t+0.5*h)
    k4 = h*f(r+k3, t+h)
    r += (k1+2*k2+2*k3+k4)/6
plot(tpoints, xpoints)
plot(tpoints, ypoints)
```
