```python
from math import pi
from numpy import linspace,array,cos,sin
#numpyのsin,cosをインポート mathではだめ。
from matplotlib.pyplot import plot
```

```python
theta = linspace(0,2*pi,1000)
```

```python
x = 2*cos(theta) + cos(2*theta)
#この部分で、mathのsin,cosは使えない
```

```python
y = 2*sin(theta) - sin(2*theta)
```

```python
plot(x,y)
```
