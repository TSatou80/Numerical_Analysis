```python
from matplotlib.pyplot import plot,xlabel, ylabel,imshow, colorbar, jet, hot, gray
from numpy import linspace,array,zeros,empty
```

```python

def mandel(c,z=0):
    for i in range(100):
        z = z**2 + c
        if abs(z)  > 2:
            break
    return abs(z)
```

```python
N  = 1000
c = empty([N,N],complex)
x = linspace(-2,2,N)
y = linspace(-2j,2j,N)
for i in range(N):
    for j in range(N):
        c[i][j] = x[i] + y[j]

z = 0
zz = empty([N,N],float)
for i in range(N):
    for j in range(N):
        zz[i][j] = mandel(c[i][j])
```

```python
imshow(zz)
colorbar()
```

```python
imshow(zz)
hot()
# colorbar()
```

```python
imshow(zz)
jet()
# colorbar()
```
