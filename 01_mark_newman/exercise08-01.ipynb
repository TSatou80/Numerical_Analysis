```python
from math import floor
from numpy import arange,linspace
from pylab import plot
```

```python
a = 0.0
b = 5
N = 100
h = (b-a)/N
RC = 0.03
tpoints = arange(a,b,h)
xpoints = []
x = 0.0
def V(t):
    if floor(2*t)%2 == 0:
        Vin = 1
    else:
        Vin = -1
    return Vin
def f(x,t):
    return (V(t) - x)/RC
for t in tpoints:
    xpoints.append(x)
    k1 = h*f(x,t)
    k2 = h*f(x+0.5*k1, t+0.5*h)
    k3 = h*f(x+0.5*k2, t+0.5*h)
    k4 = h*f(x+k3, t+h)
    x += (k1+2*k2+2*k3+k4)/6
plot(tpoints, xpoints)
```
