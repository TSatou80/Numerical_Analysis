```python
from numpy import loadtxt
from pylab import plot,show,xlim
```

```python
data = loadtxt("sunspots.txt")
```

```python
x = []
y = []
for i in range(1000):
    x.append(data[i,0])
    y.append(data[i,1])
```

```python
tmp = 0
avg = []
for i in range(100):
    for j in range(10):
        tmp += data[10*i+j,1]
    avg.append(tmp/10)
    tmp = 0
```

```python
avg_x = []
for i in range(100):
    avg_x.append(i*10)
```

```python
#平均のグラフ
plot(avg_x,avg)
xlim(0,1000)
```

```python
#元のグラフ
plot(data[:,0],data[:,1])
xlim(0,1000)
```
