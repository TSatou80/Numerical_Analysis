```python
from numpy import empty
from pylab import plot, xlabel, ylabel, show
```

```python
#constants
L = 0.01 #thickness of steel in meters
D = 4.25e-6 #thermal diffusibity
N = 100 #number of divisions in grid
a = L/N #grid spacing
h = 1e-4 # time step
epsilon = h/100

Tlo = 0.0
Tmid = 20.0
Thi = 50.0

t1 = 0.01
t2 = 0.1
t3 = 0.4
t4 = 1.0
t5 = 10.0
tend = t5 + epsilon

#create arrays
T = empty(N+1,float)
T[0] = Thi
T[N] = Tlo
T[1:N] = Tmid
Tp = empty(N+1, float)
Tp[0] = Thi
Tp[N] = Tlo

#main loop
t = 0.0
c = h*D/(a*a)
while t < tend:
    #calculate the new values of T
#     for i in range(1,N):
#         Tp[i] = T[i] + c*(T[i+1]+T[i-1]-2*T[i])
    Tp[1:N] = T[1:N] + c*(T[0:N-1] + T[2:N+1] - 2*T[1:N])
    T,Tp = Tp,T
    t += h
    
    #make plots at the given times
    if abs(t-t1) < epsilon:
        plot(T)
    if abs(t-t2) < epsilon:
        plot(T)
    if abs(t-t3) < epsilon:
        plot(T)
    if abs(t-t4) < epsilon:
        plot(T)
    if abs(t-t5) < epsilon:
        plot(T)

xlabel("X")
ylabel("T")
show()
```
